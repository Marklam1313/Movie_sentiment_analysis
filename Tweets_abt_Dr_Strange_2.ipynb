```python
count = 0

list_of_tweets = []
tweet_id = None
max_results = 10
query = 'multiverseofmadness lang:en -is:retweet'

for i in range(0,5000):
    count += 1
    if tweet_id is None:
        tweet = client.search_recent_tweets(query=query, max_results=max_results)
    else:
        tweet = client.search_recent_tweets(query=query, max_results=max_results, until_id=tweet_id)
            
    print(str(count) + ': ' + str(tweet.data[-1].id))    
    tweet_id = tweet.data[-1].id
    list_of_tweets.extend(tweet.data)
```

```python
'''
Utility function to clean tweet text by removing links, special characters
using simple regex statements.
'''
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
```

```python
def get_tweet_sentiment(text):
    result = TextBlob(text)
    if result.sentiment.polarity > 0:
        return 'positive'
    elif result.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'
```

```python
# RUN CELLS BELOW AFTER FETCHING TWEETS
```

```python
tweets = []

for data in list_of_tweets:
    parsed_tweet = {}
    
    parsed_tweet['text'] = data.text
    text = clean_tweet(data.text)
    parsed_tweet['sentiment'] = get_tweet_sentiment(text)
    tweets.append(parsed_tweet)
```

```python
df = pd.DataFrame(tweets)
```
